## **Step 1 : Initialize Parameters**
This function `initialize_parameters` initializes the weights and biases for a neural network with given layer dimensions. Weights are initialized to 0.1 and biases to 0. The `np.random.seed(3)` ensures reproducible results.

```python
def initialize_parameters(layer_dims):

  np.random.seed(3)
  parameters = {}
  L = len(layer_dims)

  for l in range(1, L):

    parameters['W' + str(l)] = np.ones((layer_dims[l-1], layer_dims[l]))*0.1
    parameters['b' + str(l)] = np.zeros((layer_dims[l], 1))


  return parameters
```

```python
initialize_parameters([2,2,1])
```

The `linear_forward` function computes the linear part of a layer's forward propagation. It takes the activation from the previous layer (`A_prev`), the weights (`W`), and biases (`b`) for the current layer, and returns `Z = W.T * A_prev + b`.

```python
def linear_forward(A_prev, W, b):
  Z = np.dot(W.T, A_prev) + b
  return Z
```

## **Step 2 : Forward Propagation**
The `L_layer_forward` function performs the full forward pass through the neural network. It iterates through each layer, applying the `linear_forward` calculation. The output `A` (activation) from one layer becomes `A_prev` for the next.

```python
# Forward Prop
def L_layer_forward(X, parameters):

  A = X
  L = len(parameters) // 2

  for l in range(1, L+1):
    A_prev = A
    Wl = parameters['W' + str(l)]
    bl = parameters['b' + str(l)]

    A = linear_forward(A_prev, Wl, bl)

  return A,A_prev
```

## **Step 3 : Weights and Bias Updation**
The `update_parameters` function implements the backpropagation step to adjust the weights and biases. It calculates the gradients based on the difference between the actual output (`y`) and the predicted output (`y_hat`), and updates the parameters using a learning rate (0.001 in this case). This specific implementation shows a manual, hardcoded gradient update for a 2-layer network.

```python
def update_parameters(parameters,y,y_hat,A1,X):
  parameters['W2'][0][0] = parameters['W2'][0][0] + (0.001 * 2 * (y - y_hat)*A1[0][0])
  parameters['W2'][1][0] = parameters['W2'][1][0] + (0.001 * 2 * (y - y_hat)*A1[1][0])
  parameters['b2'][0][0] = parameters['W2'][1][0] + (0.001 * 2 * (y - y_hat))

  parameters['W1'][0][0] = parameters['W1'][0][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0]*X[0][0])
  parameters['W1'][0][1] = parameters['W1'][0][1] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0]*X[1][0])
  parameters['b1'][0][0] = parameters['b1'][0][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0])

  parameters['W1'][1][0] = parameters['W1'][1][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0]*X[0][0])
  parameters['W1'][1][1] = parameters['W1'][1][1] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0]*X[1][0])
  parameters['b1'][1][0] = parameters['b1'][1][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0])
```

This block implements the training loop for the neural network. It iterates for a specified number of `epochs`. In each epoch, it processes every data point in the `df` DataFrame:

1.  **Input and Output Extraction:** Extracts features (`X`) and target (`y`) for the current data point.
2.  **Forward Propagation:** Calls `L_layer_forward` to get the predicted output (`y_hat`) and intermediate activation (`A1`).
3.  **Parameter Update:** Calls `update_parameters` to adjust the network's weights and biases based on the prediction error.
4.  **Loss Calculation:** Calculates the squared error loss for the current data point and appends it to a list.

After each epoch, it prints the average loss for that epoch. Finally, it prints the updated `parameters`.

```python
# epochs implementation

parameters = initialize_parameters([2,2,1])
epochs = 5

for i in range(epochs):

  Loss = []

  for j in range(df.shape[0]):

    X = df[['cgpa', 'profile_score']].values[j].reshape(2,1) # Shape(no of features, no. of training example)
    y = df[['lpa']].values[j][0]

    # Parameter initialization


    y_hat,A1 = L_layer_forward(X,parameters)
    y_hat = y_hat[0][0]

    update_parameters(parameters,y,y_hat,A1,X)

    Loss.append((y-y_hat)**2)

  print('Epoch - ',i+1,'Loss - ',np.array(Loss).mean())

parameters
```
